グレースケール変換

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
img_gray.shape
```

フィルタをかける

```python
kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```python
img_conv = cv2.filter2D(img_gray, -1, kernel)
plt.imshow(img_conv, cmap='gray')
```

```python
kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])

img_conv = cv2.filter2D(img_gray, -1, kernel)
plt.imshow(img_conv, cmap='gray')
```
